```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.figsize'] = (8,5.5)
```

```python
states2 = pd.read_csv('states_all2.csv')
states2.info()
```

```python
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.boxplot(states2["TOTAL_REVENUE"])
plt.title("TOTAL_REVENUE kutu grafiği (whis=1.5)", fontdict=baslik_font)

plt.subplot(1,2,2)
plt.boxplot(states2["TOTAL_EXPENDITURE"])
plt.title("TOTAL_EXPENDITURE kutu grafiği (whis=1.5)", fontdict=baslik_font)

plt.show()
```

```python
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
eksen_font  = {'family': 'arial', 'color': 'darkblue','weight': 'bold','size': 10 }

plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.boxplot(states2["TOTAL_REVENUE"],whis=1.5)
plt.title("TOTAL_REVENUE kutu grafiği (whis=1.5)", fontdict=baslik_font)

plt.subplot(1,2,2)
plt.boxplot(states2["TOTAL_EXPENDITURE"],whis=1.5)
plt.title("TOTAL_EXPENDITURE kutu grafiği (whis=1.5)", fontdict=baslik_font)

plt.show()
```

2. soru

```python
from scipy.stats.mstats import winsorize

winsorize_states2_TOTALREVENUE = winsorize(states2["TOTAL_REVENUE"], (0, 0.10))

winsorize_states2_TOTALREVENUE
```

```python
plt.boxplot(winsorize_states2_TOTALREVENUE)
plt.title("TOTAL REVENUE kutu grafiği (whis=1.5)", fontdict=baslik_font)
plt.show()
```

```python
from scipy.stats.mstats import winsorize

winsorize_states2_TOTALEXPENDITURE = winsorize(states2["TOTAL_EXPENDITURE"], (0, 0.10))

winsorize_states2_TOTALEXPENDITURE
```

```python
plt.boxplot(winsorize_states2_TOTALEXPENDITURE)
plt.title("TOTAL EXPENDITURE kutu grafiği (whis=1.5)", fontdict=baslik_font)
plt.show()
```

3. soru

```python
states2["BUDGET_DEFICIT"]=states2["TOTAL_REVENUE"]-states2["TOTAL_EXPENDITURE"]
```

```python
plt.boxplot(states2["BUDGET_DEFICIT"])
plt.title("BUDGET DEFICIT kutu grafiği (whis=1.5)", fontdict=baslik_font)
```

```python
from scipy.stats.mstats import winsorize

winsorize_states2_BUDGETDEFICIT = winsorize(states2["BUDGET_DEFICIT"], (0.05, 0.10))
plt.boxplot(winsorize_states2_BUDGETDEFICIT)
plt.title("BUDGET DEFICIT kutu grafiği (whis=1.5)", fontdict=baslik_font)
plt.show()
```

4. soru

```python
states2["BUDGET_DEFICIT_WINS"]=winsorize_states2_TOTALREVENUE-winsorize_states2_TOTALEXPENDITURE
plt.boxplot(states2["BUDGET_DEFICIT_WINS"])
plt.title("BUDGET DEFICIT WINSORIZE kutu grafiği (whis=1.5)", fontdict=baslik_font)
```

```python
from scipy.stats.mstats import winsorize

winsorize_states2_BUDGETDEFICITWINS = winsorize(states2["BUDGET_DEFICIT_WINS"], (0.03, 0.05))
plt.boxplot(winsorize_states2_BUDGETDEFICITWINS)
plt.title("BUDGET DEFICIT WINSORIZE kutu grafiği (whis=1.5)", fontdict=baslik_font)
plt.show()
```

5. soru

```python
plt.figure(figsize = (15,5))
winsorize_states2_BUDGETDEFICIT = winsorize(states2["BUDGET_DEFICIT"], (0.05, 0.10))
plt.subplot(1,2,1)
plt.boxplot(winsorize_states2_BUDGETDEFICIT)
plt.title("BUDGET DEFICIT kutu grafiği (whis=1.5)", fontdict=baslik_font)

winsorize_states2_BUDGETDEFICITWINS = winsorize(states2["BUDGET_DEFICIT_WINS"], (0.03, 0.05))
plt.subplot(1,2,2)
plt.boxplot(winsorize_states2_BUDGETDEFICITWINS)
plt.title("BUDGET DEFICIT WINSORIZE kutu grafiği (whis=1.5)", fontdict=baslik_font)
plt.show()
```

```python
plt.figure()

plt.subplot(121)
plt.hist(winsorize_states2_BUDGETDEFICIT)
plt.subplot(122)
plt.hist(winsorize_states2_BUDGETDEFICITWINS)
plt.show()
```
